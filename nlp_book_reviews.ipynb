## Text Analysis with NLP

## Sentiment analysis

The dataset for the sentiment analysis is reviews of "Harry Potter and the Sorcerer's Stone ((Harry Potter #1))" by J.K. Rowling taken from a book review website "goodreads": https://www.goodreads.com/book/show/3.Harry_Potter_and_the_Sorcerer_s_Stone

24 reviews were copied from the website and stored in a csv file, with an effort to include some reviews made by reviewers who had also reviewed Harry Potter and the Chamber of Secrets (Harry Potter #2) for the similarity analysis.

```python
import pandas as pd 
book_1= pd.read_csv('https://raw.githubusercontent.com/e-hendry/datasets/main/harrypotter_reviews_book1.csv',encoding="UTF-8")
book_1
```

### Sentiment Analysis with TextBlob’s Default Sentiment Analyzer
We can start by using TextBlob's default senitment analyzer to analyze the reviews. TextBlob's default sentiment analyzer returns a polarity and subjectivity value. The polarity value is float between -1 and 1 where 1 is the most positive statement and -1 is the most negative statement. Subjectivity represents how muuch personal opinion, emotion or judgement is included in the text and is returned as a flat between 0 and 1 where 1 means that statement contains mostly opinion, and 0 means that statement contains mostly factual information (i.e. objective).

```python
from textblob import TextBlob
book_1['polarity'] = book_1['review'].apply(lambda text: TextBlob(text).sentiment.polarity)
book_1['subjectivity'] = book_1['review'].apply(lambda text: TextBlob(text).sentiment.subjectivity)
book_1
```

We can evaluate how well the default sentiment analyzer worked by comparing the number of stars to the calculated polarity value. We would expect the lowest polarity value for a one star review and the highest polarity value for a five start review. As shown by the figure below, we have low polarity readings for one star reviews. The polarity values are higher for two and three star reviews, however there is less of a difference between the polarity value of two and three star reviews and one star reviews. There is a significantly larger spread in the polarity values for four and five star reviews, with the lowest polarity value for four stars being lower than the lowest value for two and three stars, and the lowest value for five stars being lower than the value for two, three and four stars. This suggests that the default sentiment anlyzer was not as effective at detecting the sentiment of positive reviews (those with four or five stars). 

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.boxenplot(x='star_rating', y='polarity', data=book_1)
plt.show()
```

Next, we can try the Naive Bayes Analyzer which returns the sentiment (positive or negative), a positive percentage and a negative percentage. The original star ratings can be converted to a percentage value using the following scale: 
- 1 star: 0% - 20% 
- 2 stars: 20% - 40%
- 3 stars: 40% - 60% 
- 4 stars: 60% - 80% 
- 5 stars: 80% - 100%

```python
from textblob.sentiments import NaiveBayesAnalyzer
from textblob import Blobber
```

```python
tb = Blobber(analyzer=NaiveBayesAnalyzer())
```

```python
book_1[['sentiment', 'pos', 'neg']] = book_1['review'].apply(lambda v: pd.Series(tb(v).sentiment))
book_1['pos_to_star'] = [1 if x <0.20 else 2 if x<0.40 else 3 if x<0.60 else 4 if x<0.80 else 5 for x in book_1['pos']] 
book_1['correct_star_calc'] = (book_1['pos_to_star']==book_1['star_rating']).astype(int)
book_1
```

### Score for Sentiment Analysis

```python
total_correct = sum(book_1['correct_star_calc'])
total_values = book_1['correct_star_calc'].count()
percentage_correct = total_correct/total_values
print(f'percentage of star ratings correctly classified: {percentage_correct:.2%}')
```

As shown above, the calculated score for how closely the analysis matched the book ratings is 41.67%. 

### Confusion Matrix

The confusion matrix below summarizes the predicted star ratings compared with the actual star ratings. Based on the matrix, the values were classified as follows:

- One-star: 1 correctly classified, 2 mis-classified as a 5 
- Two stars: 0 correctly classified, 1 mis-classified as a 3, 1 mis-classified as a 5
- Three-stars: 1 correctly classified, 1 mis-classified as 4, 1 mis-classified as 5 
- Four-stars: 0 correctly classified, 1 mis-classified as a 3 and 3 mis-classified as a 5 
- Five-stars: 8 correctly classified, 2 mis-classified as a 3, 1 mis-classified as a 2 and 1 mis-classified as a 1 

```python
from sklearn.metrics import confusion_matrix


expected=book_1['star_rating']
predicted=book_1['pos_to_star']
confusion = confusion_matrix(y_true=expected, y_pred=predicted)

print('Confusion matrix:')
print(confusion)
```

```python
import pandas as pd
import seaborn as sns

confusion_df = pd.DataFrame(confusion, index=range(5), columns=range(5))
axes = sns.heatmap(confusion_df, annot=True, cmap='nipy_spectral_r')
```

## Similarity analysis

The dataset for the similarity analysis is reviews of "Harry Potter and the Chamber of Secrets ((Harry Potter #2))" by J.K. Rowling taken from a book review website "goodreads": https://www.goodreads.com/book/show/15881.Harry_Potter_and_the_Chamber_of_Secrets. 

17 of the 24 reviews in this dataset are written by the same reviewers who wrote a review for Harry Potter #1 that is included in the first dataset.

```python
import pandas as pd 
book_2= pd.read_csv('https://raw.githubusercontent.com/e-hendry/datasets/main/harrypotter_reviews_book2.csv',encoding="UTF-8")
book_2
```

```python
import spacy
nlp = spacy.load('en_core_web_lg') 
```

```python
result=[]

reviewers_1=book_1['reviewer_name']
reviews_1=book_1['review']
reviewers_2=book_2['reviewer_name']
reviews_2=book_2['review']

for i in range(0,len(reviewers_1)):
    current_reviewer_1=reviewers_1[i]
    current_review_1=reviews_1[i]
    for x in range(0,len(reviewers_2)):
        result_i=[]
        current_reviewer_2=reviewers_2[x]
        if current_reviewer_1==current_reviewer_2:
            result_i.append(1)
        else: 
            result_i.append(0)
        current_review_2=reviews_2[x]
        doc_rev1=nlp(current_review_1)
        doc_rev2=nlp(current_review_2)
        sim_result = doc_rev1.similarity(doc_rev2)
        #result_i.append(current_reviewer_1)
        #result_i.append(current_reviewer_2)
        result_i.append(sim_result)
        result.append(result_i)
```

```python
count_same=0
count_diff=0
total_same=0
total_diff=0

for i in result: 
    if i[0]==1:
        count_same+=1
        total_same+=i[1]
    else: 
        count_diff+=1
        total_diff+=i[1]
        

avg_same = total_same/count_same
avg_diff = total_diff/count_diff
        
print(f'Total combinations where review was written by the same reviewer: {count_same}')
print(f'Average similarity value for reviews written by the same reviewer: {avg_same:.2f}')

print(f'Total combinations where review was written by a different reviewer: {count_diff}')
print(f'Average similarity value for reviews written by a different reviewer: {avg_diff:.2f}')
```

Based on the data shown above and the box plots below, there is not a large difference between the average similarity value calculated for reviews written by the same reviewer and those written by different reviwers. The average similarity value for reviews written by the same reviewer was 0.78 and the average similarity value for reviews written by different reviewers was 0.73. This suggests that the analysis performed did not accurately determine which reviews were written by the same reviewer. 

Given that all these reviews are written about the same topic (Harry Potter books) 0.8 seems a reasonable similarity value for to indicate that the reviews were written by the same reviewer. If we consider any reviews that produced a similary value of 0.8 or higher to be an indication that they were written by the same reviewer, and a similarity value lower than 0.8 to indicate that they were written by a different reviewer, we can calculate an approximate score for the similarity analysis.

```python
count_good=0
count_bad=0

for i in result: 
    if i[0]==1 and i[1]>=0.8:
        count_good+=1
    elif i[0]==0 and i[1]<0.8:
        count_good+=1
    else: 
        count_bad+=1

percentage_good = count_good/(count_good+count_bad)        
print(f'percentage of similarity scores that accurately reflected review authorship: {percentage_good:.2%}')
```

```python
same_reviewer=[]
similarity=[]

for i in result: 
    same_reviewer.append(i[0])
    similarity.append(i[1])                        
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.boxenplot(x=same_reviewer, y=similarity)
plt.show()
```
