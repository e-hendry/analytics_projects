### Dimensionality Reduction using TSNE Estimator

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=11) #data reduction object 
reduced_wine_tsne = tsne.fit_transform(wine.data) #do a fit transform

print(f'Reduced data shape: reduced_wine.shape = {reduced_wine_tsne.shape}') #we have reduced the data down to two dimensions
```

```python
from sklearn.manifold import TSNE


%matplotlib inline
import matplotlib.pyplot as plt

dots = plt.scatter(reduced_wine_tsne[:, 0], reduced_wine_tsne[:, 1], c='black')
```

### Dimensionality Reduction Using Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=11)  # reduce to two components
pca.fit(wine.data)
```

```python
reduced_wine_pca = pca.transform(wine.data)
reduced_wine_pca.shape
```

```python
dots = plt.scatter(reduced_wine_pca[:, 0], reduced_wine_pca[:, 1], c='black')
```

Based on the visualizations shown above for data reduced using PCA and TSNE, PCA appears to produce the best data for cluster analysis. We would expect to see 3 clusters in the data because we have 3 classes of wine, however in the visualization for the data reduced by PCA we can see two non-clearly defined clusters. The visualization for the data reduced using PCA suggests that we can try performing the k-means clustering with two clusters, and then perform the clustering with three clusters, as we have 3 different class types in this dataset.

### K-means Clustering with Two Clusters

#### Creating a k-means estimator and fitting the model

```python
from sklearn.cluster import KMeans

kmeans_2c = KMeans(n_clusters=2, random_state=11) 
kmeans_2c.fit(wine.data)
```

```python
wine_df['predicted_classes_2c']=kmeans_2c.labels_
```

```python
reduced_wine_pca_df = pd.DataFrame(reduced_wine_pca, columns=['Component 1', 'Component 2'])
reduced_wine_pca_df['class'] = wine_df['class']
```

#### Visualizing the Results of the k-means clustering

```python
import matplotlib.pyplot as plt

axes = sns.scatterplot(data=reduced_wine_pca_df, hue=wine_df['predicted_classes_2c'], legend='brief', 
                       x='Component 1', y='Component 2') 

wine_centers = pca.transform(kmeans_2c.cluster_centers_)
dots = plt.scatter(wine_centers[:,0], wine_centers[:,1], s=100, c='k')
```

As there are three classes in the dataset, but this clustering assigned labels to two clusters, we cannot compare the actual and predicted labels.

###  K-means Clustering with Three Clusters

#### Creating a k-means estimator and fitting the model

```python
from sklearn.cluster import KMeans

kmeans_3c = KMeans(n_clusters=3, random_state=11) 
kmeans_3c.fit(wine.data)
```

```python
wine_df['predicted_classes_3c']=kmeans_3c.labels_
```

#### Visualizing the Results of the k-means clustering

```python
axes = sns.scatterplot(data=reduced_wine_pca_df, hue=wine_df['predicted_classes_3c'], legend='brief', 
                       x='Component 1', y='Component 2') 

wine_centers = pca.transform(kmeans_3c.cluster_centers_)
dots = plt.scatter(wine_centers[:,0], wine_centers[:,1], s=100, c='k')
```

The graph above shows the results of the k-means clustering using three clusters. We can compare this to a visualization of the reduced data to identify which labels refer to which classes.

```python
axes = sns.scatterplot(data=reduced_wine_pca_df, hue='class', legend='brief', 
                       x='Component 1', y='Component 2') 

wine_centers = pca.transform(kmeans_3c.cluster_centers_)
dots = plt.scatter(wine_centers[:,0], wine_centers[:,1], s=100, c='k')
```

As shown by the labelled data above, the cluster on the left is class 1, the middle cluster is class 2 and the class on the right is class 0. This means that in our newly labelled data, label 0 corresponds to class 2, label 1 corresponds to class 0 and label 2 corresponds to class 1. We can use this information to calculate the percentage of values that were assigned correctly in each class.

#### Comparing the Assigned Labels with the Actual Labels

```python
def kmeans_stats(selected_class):

    actual_class=wine_df['class']
    predicted_class=wine_df['predicted_classes_3c']

    count_correct_0=0
    count_correct_1=0
    count_correct_2=0
    class_0_count=0
    class_1_count=0
    class_2_count=0

    for i in range(0,len(actual_class)):
        if selected_class=='class_0' and actual_class[i]==selected_class:
            class_0_count+=1
            if actual_class[i]==selected_class and predicted_class[i]==1:
                count_correct_0+=1
            percentage_correct_0=count_correct_0/class_0_count
            percentage_incorrect_0=(class_0_count-count_correct_0)/class_0_count
     
        if selected_class=='class_1' and actual_class[i]==selected_class:
            class_1_count+=1
            if actual_class[i]==selected_class and predicted_class[i]==0:
                count_correct_1+=1    
            percentage_correct_1=count_correct_1/class_1_count
            percentage_incorrect_1=(class_1_count-count_correct_1)/class_1_count
  
                
        if selected_class=='class_2' and actual_class[i]==selected_class:
            class_2_count+=1
            if actual_class[i]==selected_class and predicted_class[i]==2:
                count_correct_2+=1
            percentage_correct_2=count_correct_2/class_2_count
            percentage_incorrect_2=(class_2_count-count_correct_2)/class_2_count

        
    if selected_class=='class_0':
        print(f'Total correct predictions for {selected_class}: {count_correct_0}')
        print(f'Percentage of correct predictions for {selected_class}: {percentage_correct_0:.2%}')
        print(f'Percentage of incorrect predictions for {selected_class}: {percentage_incorrect_0:.2%}')
        print('\n')   
    if selected_class=='class_1':
        print(f'Total correct predictions for {selected_class}: {count_correct_1}')
        print(f'Percentage of correct predictions for {selected_class}: {percentage_correct_1:.2%}')
        print(f'Percentage of incorrect predictions for {selected_class}: {percentage_incorrect_1:.2%}')
        print('\n') 
    if selected_class=='class_2':
        print(f'Total correct predictions for {selected_class}: {count_correct_2}')
        print(f'Percentage of correct predictions for {selected_class}: {percentage_correct_2:.2%}')
        print(f'Percentage of incorrect predictions for {selected_class}: {percentage_incorrect_2:.2%}')
        print('\n') 
    
    return None
```

```python
kmeans_stats('class_0')
kmeans_stats('class_1')
kmeans_stats('class_2')
```

As shown by the stats above, the most accurate predictions were for class 0, where 77.97% of the data points in that cluster had the correct label. It is worth nothing that the data set used for this analysis did not appear to have obvious clusters, and that alternate algorithms should be explored to see if the results can be improved.

## **Supervised Machine Learning**

The data that was labelled using k-means clustering with three clusters will be used for the next part of this assignment.

### Splitting the Newly Labelled Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
     wine.data, wine_df['predicted_classes_3c'], random_state=11) 

print(f'X_train.shape = {X_train.shape}') #training data
print(f'X_test.shape  = {X_test.shape}') #testing data

print()
print(f'y_train.shape = {y_train.shape}') #target training data
print(f'y_test.shape  = {y_test.shape}') #target testing data 
```

### Performing the K Nearest Neighbours Classification

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X=X_train, y=y_train)
```

### Testing the Model by Predicting Wine Classes

```python
predicted = knn.predict(X=X_test) #pass in the test data
expected = list(y_test) #targets of the labels 

print('First twenty predictions:')
print(f'predicted[:20] = {predicted[:20]}')
print(f'expected[:20] = {expected[:20]}')
```

### Model Prediction Accuracy

```python
wrong = [ (pred, exp) 
          for (pred, exp) in zip(predicted, expected) 
          if pred != exp
        ]

print('Wrong predictions:')
print(wrong) 
```

```python
print(f'Prediction accuracy score = {knn.score(X_test, y_test):.2%}')
```

As shown above, the model is able to predict wine class with 100% accuracy.

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_true=expected, y_pred=predicted)

print('Confusion matrix:')
print(confusion)
```

```python
import pandas as pd
import seaborn as sns

confusion_df = pd.DataFrame(confusion, index=range(3), columns=range(3))
axes = sns.heatmap(confusion_df, annot=True, cmap='nipy_spectral_r')
```

### Classification Report

```python
from sklearn.metrics import classification_report

names = [str(wine) for wine in wine.target_names]
print(classification_report(expected, predicted, target_names=names))
```

### Comparison of Predicted Labels to Original Labels

```python
original_index=y_test.index
original_labels=wine_df['class']

predicted_values=[]

for x in range(0,len(original_index)):
    predicted_value=predicted[x]
    predicted_values.append(predicted_value)

original_values=[]
    
for i in original_index:
    original_value=original_labels[i]
    original_values.append(original_value)
 
incorrect_predictions=[]
    
for i in range(0,len(predicted_values)):
    original_val_current=original_values[i]
    predicted_val_current=predicted_values[i]
    if original_val_current=='class_0' and predicted_val_current!=1:
        #print(f'expected: 1, predicted: {predicted_val_current}')
        incorrect_predictions.append(f'expected: 1, predicted: {predicted_val_current}')
    if original_val_current=='class_1' and predicted_val_current!=0:
        #print(f'expected: 2, predicted: {predicted_val_current}') 
        incorrect_predictions.append(f'expected: 2, predicted: {predicted_val_current}')
    elif original_val_current=='class_2' and predicted_val_current!=2:
        incorrect_predictions.append(f'expected: 0, predicted: {predicted_val_current}')
        #print(f'expected: 0, predicted: {predicted_val_current}')
        

print(f'Total number of incorrect predictions is: {len(incorrect_predictions)}')   
print(incorrect_predictions)
print(f'Actual model accuracy: {(len(predicted_values)-len(incorrect_predictions))/len(predicted_values):.2%}')
```

As shown above, when compared to the original label values the actual model accuracy is 77.78%, which is significantly lower than the value previously calculated.
